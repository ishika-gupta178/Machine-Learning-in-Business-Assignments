# Importing Libraries

```python
import numpy as np
import pandas as pd

# to convert categorical values to numeric
from sklearn.preprocessing import OneHotEncoder

# to balance the dataset
from imblearn.over_sampling import SMOTE

# to create decision tree and random forest models
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# to evaluate the performance
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# to perform hyperparameter tuning
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV

# to perform cross validation and see how well the model generalized and can predict unseen data accurately
from sklearn.model_selection import cross_val_score
```

# Reading data into pandas dataframe

```python
train_df = pd.read_csv("Insurance Fraud - TRAIN-3000.csv")
train_df
```

```python
test_df = pd.read_csv("Insurance Fraud -TEST-12900.csv")
test_df
```

# Data Preprocessing

```python
train_index = train_df.index
train_index
```

```python
test_df.index
```

```python
# combining the training and testing datasets together and then one hot encoding it and then splitting into training and testing
# because the testing dataset has some values which are not in the training dataset

# if we first split train and test and then one hot encode them individually then we get different columns for train and test
# we cannot have different columns for train and test datasets after encoding it

combined_df = pd.concat([train_df,test_df])
combined_df
```

```python
# checking class distribution
# dataset is imbalanced
combined_df["FRAUDFOUND"].value_counts()
```

```python
# separating input and output
combined_df_X = combined_df.drop("FRAUDFOUND", axis = 1)
combined_df_Y = combined_df[["FRAUDFOUND"]]
```

```python
# performing one hot encoding on categorical columns
categorical_columns = []

for col in combined_df_X.columns:
  if combined_df_X[col].dtype=='O':
    categorical_columns.append(col)

ohe = OneHotEncoder(handle_unknown='ignore',sparse=False)
combined_Xcat = pd.DataFrame(ohe.fit_transform(combined_df[categorical_columns]),
                             columns=ohe.get_feature_names_out(),
                             index=combined_df_X.index)
combined_df_X = pd.concat([combined_df_X,combined_Xcat],axis=1)
combined_df_X.drop(labels = categorical_columns,axis=1,inplace=True)
combined_df_X
```

```python
# splitting dataset into xtrain, ytrain, xtest and ytest
X_train = combined_df_X.iloc[0:2999,:]
```

```python
X_train
```

```python
X_test = combined_df_X.iloc[2999:,:]
```

```python
X_test
```

```python
y_train = combined_df_Y.iloc[0:2999,:]
y_train
```

```python
y_test = combined_df_Y.iloc[2999:,:]
y_test
```

```python
# sampling the data to balance the dataset using SMOTE

sm = SMOTE(random_state=19)

X_resampled, y_resampled = sm.fit_resample(X_train, y_train)
X_resampled
```

```python
y_resampled
```

```python
# checking class distribution after sampling
# dataset is now balanced
y_resampled.value_counts()
```

# Models

## Decision Tree - No hyperparameter tuning

```python
dt = DecisionTreeClassifier()
dt.fit(X_resampled, y_resampled)
```

```python
train_pred = dt.predict(X_resampled)
test_pred = dt.predict(X_test)

#Model Accuracy
print("Train Accuracy:", accuracy_score(y_resampled, train_pred))
print("Test Accuracy:", accuracy_score(y_test, test_pred))
print("Confusion Matrix for Decision Tree:")
print(confusion_matrix(y_test, test_pred))
print("Max Depth",dt.get_depth())
print("Leaf",dt.get_n_leaves())
print('Printing the precision and recall, among other metrics')
print(classification_report(y_test, test_pred))
```

## Random Forest - No hyperparameter tuning

```python
rf = RandomForestClassifier(random_state = 20)
rf.fit(X_resampled, y_resampled)
```

```python
train_pred_1 = rf.predict(X_resampled)
test_pred_1 = rf.predict(X_test)

#Model Accuracy
print("Train Accuracy:", accuracy_score(y_resampled, train_pred_1))
print("Test Accuracy:", accuracy_score(y_test, test_pred_1))
print("Confusion Matrix for Decision Tree:")
print(confusion_matrix(y_test, test_pred_1))
print('Printing the precision and recall, among other metrics')
print(classification_report(y_test, test_pred_1))
```

## Decision Tree - Hyperparameter tuning using Random Search

```python
#Hyperparameter tuning done for decision tree classifier

#RANDOM SEARCH

import time
start_time = time.time()

print("RandomizedSearchCV-Decision tree")

parameters={
            'max_depth': range(5,50,5),
            'criterion':['gini','entropy'],
            # 'max_leaf_nodes':range(10,500,10),
            'min_samples_split': range(5,50,3),
           }

dt_random = RandomizedSearchCV(dt,parameters,n_iter=25,cv=5)

dt_random.fit(X_resampled, y_resampled)

grid_parm = dt_random.best_params_
print(grid_parm)

train_pred_2 = dt_random.predict(X_resampled)
test_pred_2 = dt_random.predict(X_test)

print("Train Accuracy:", dt_random.score(X_resampled, y_resampled))
print("Test Accuracy:", dt_random.score(X_test, y_test))

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
dtRand = DecisionTreeClassifier(**grid_parm)
dtRand.fit(X_resampled,y_resampled)

dtRand_predict_train = dtRand.predict(X_resampled)
dtRand_predict = dtRand.predict(X_test)
print("Train Accuracy:", accuracy_score(y_resampled,dtRand_predict_train))
print("Test Accuracy:", accuracy_score(y_test,dtRand_predict))
print("Confusion Matrix for Decision Tree:")
print(confusion_matrix(y_test,dtRand_predict))
print('Printing the precision and recall, among other metrics')
print(classification_report(y_test, dtRand_predict))
clf_cv_score = cross_val_score(dtRand, X_resampled, y_resampled, cv=5, scoring="balanced_accuracy")
print(clf_cv_score)
```

## Decision Tree - Hyperparameter tuning using Grid Search

```python
import time
start_time = time.time()

print("GridSearchCV-Decision tree")
dt_grid = GridSearchCV(dt,parameters,cv = 5)

dt_grid.fit(X_resampled, y_resampled)

grid_parm1=dt_grid.best_params_

print(grid_parm1)


train_pred_3 = dt_grid.predict(X_resampled)
test_pred_3 = dt_grid.predict(X_test)

print("Train Accuracy:", dt_grid.score(X_resampled, y_resampled))
print("Test Accuracy:", dt_grid.score(X_test, y_test))

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
dtGrid = DecisionTreeClassifier(**grid_parm1)
dtGrid.fit(X_resampled,y_resampled)

dtGrid_predict_train = dtGrid.predict(X_resampled)
dtGrid_predict = dtGrid.predict(X_test)
print("Train Accuracy:", accuracy_score(y_resampled,dtGrid_predict_train))
print("Test Accuracy:", accuracy_score(y_test,dtGrid_predict))
print("Confusion Matrix for Decision Tree:")
print(confusion_matrix(y_test,dtGrid_predict))
print('Printing the precision and recall, among other metrics')
print(classification_report(y_test, dtGrid_predict))
clf_cv_score1 = cross_val_score(dtGrid, X_resampled, y_resampled, cv=5, scoring="balanced_accuracy")
print(clf_cv_score1)
```

```python
rfRand = RandomForestClassifier(**grid_parm_rf)
rfRand.fit(X_resampled,y_resampled)

rfRand_predict_train = rfRand.predict(X_resampled)
rfRand_predict = rfRand.predict(X_test)
print("Train Accuracy:", accuracy_score(y_resampled,rfRand_predict_train))
print("Test Accuracy:", accuracy_score(y_test,rfRand_predict))
print("Confusion Matrix for Decision Tree:")
print(confusion_matrix(y_test,rfRand_predict))
print('Printing the precision and recall, among other metrics')
print(classification_report(y_test, rfRand_predict))
clf_cv_score = cross_val_score(rfRand, X_resampled, y_resampled, cv=5, scoring="balanced_accuracy")
print(clf_cv_score)
```

```python
# accuracy can be misleading: model can have a high accuracy but that is because it is classifying one class accurately and the other
# class not very accurately.

# we focus on precision and recall
# for all models, for the 'yes' class (fraudulent transactions, minority class), we get high recall and low precision
# this means that our model is not being picky about which transactions are fradulent
# its following the 'better be safe than sorry' strategy.
# its classifying a lot of transactions as fradulent. this will cover most of the transactions that are actually fraudelent, but it
# also has a lot of false positives. (missclassifying non-fradulent transactions) as fraudelent.
# the penalty for mislabeling a legitimate transaction as fraud is having the user frozen out of their finances and unable to make
# payments. but this also prevents money from being stolen.
```
