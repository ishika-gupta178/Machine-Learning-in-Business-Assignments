# Importing libraries

```python

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import RandomizedSearchCV

from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, log_loss


from imblearn.over_sampling import SMOTE

from collections import Counter

from vecstack import stacking
```

# Reading Kaggle Datasets

```python
# reading the kaggle training dataset

train_data = pd.read_csv("RevisedHomesiteTrain1.csv")
train_data
```

```python
# reading the kaggle testing dataset

test_data = pd.read_csv("RevisedHomesiteTest1.csv")
test_data
```

# Data Preprocessing

```python
# dropping the last column from the kaggle test dataset because that column is not present in the kaggle training dataset
test_data = test_data.drop('GeographicField64', axis=1)
test_data
```

```python
train_data.info()

# dataset has all numerical type columns only
```

```python
train_data.describe()
```

```python
# checking if there are any missing values in the dataset
if 1 in train_data.isnull().sum().values:
  print("missing value")

# there are no missing values in the dataset
```

```python
train_data['QuoteConversion_Flag'].value_counts()

# dataset is imbalanced
```

```python
# Separating the input and output of the kaggle training dataset
X = train_data.iloc[:,:-1]
y = train_data['QuoteConversion_Flag']
```

```python
# creating new records belonging to the minority class using SMOTE. (performing oversampling to balance the dataset)
# making the number of minority class records equal to the majority class records
# 1 is the minority class and 0 is the majority class

print('Original dataset shape %s' % Counter(y))
sm = SMOTE(sampling_strategy={1: 52738})
X_res, y_res = sm.fit_resample(X, y)
print('Resampled dataset shape %s' % Counter(y_res))
```

```python
# splitting the kaggle training dataset into training and testing dataset to measure performance
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, random_state=42)
```

```python
# distribution of the class in the training dataset
y_train.value_counts()
```

# Building and Training models

## Neural Network - Multi layered perceptron

```python

mlp = MLPClassifier(max_iter = 100)
mlp.fit(X_train, y_train)
mlp_predict=mlp.predict(X_test)
print("accuracy Score (training) for MultiLayer Perceptron:{0:6f}".format(mlp.score(X_test,y_test)))
print("Confusion Matrix for MultiLayer Perceptron:")
print(confusion_matrix(y_test,mlp_predict))
print(classification_report(mlp_predict, y_test))
```

neural networks perform well with large datasets because of their deep architecture. it allows them to capture intricate patterns among a large number of features.

## SVC - Support Vector Classifier

```python
svm = SVC(max_iter = 500)
svm.fit(X_train, y_train)
svm_predict=svm.predict(X_test)
print("accuracy Score (training) for SVM Classifier:{0:6f}".format(svm.score(X_test,y_test)))
print("Confusion Matrix for SVM Classifier:")
print(confusion_matrix(y_test,svm_predict))
print(classification_report(svm_predict, y_test))
```

SVC performs classification by creating a hyperplane that separates out the points belonging to different classes. This hyperplane is represented by the equation wx+b where w is the weight, x is the vector and b is the bias. its goal is to maximize the margin between the classes and minimize classification error by computing the pairwise similarities between the data points or vectors.  The larger our dataset, i.e, the more number of vectors we have, the more computationally expensive it becomes. This is why SVM does not perform well on large datasets.

## Decision Tree

```python
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
clf_predict=clf.predict(X_test)
print("accuracy Score (training) for Decision Tree:{0:6f}".format(clf.score(X_test,y_test)))
print("Confusion Matrix for Decision Tree")
print(confusion_matrix(y_test,clf_predict))
print(classification_report(clf_predict, y_test))
```

decision trees work well for large datasets with large number of features because decision trees inherently perform feature selection. They can understand which features are important and which are not. They can capture intricate patterns between the features as well

## Random Forest

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
rfc_predict=rfc.predict(X_test)
print("accuracy Score (training) for RandomForest:{0:6f}".format(rfc.score(X_test,y_test)))
print("Confusion Matrix for Random Forest:")
print(confusion_matrix(y_test,rfc_predict))
print(classification_report(rfc_predict, y_test))
```

Random forests work well because they create multiple decision trees and decrease the overall error rate. this is why random forests perform better than decision trees

## K nearest neighbors - KNN

```python
neigh = KNeighborsClassifier(n_neighbors=3)

neigh.fit(X_train, y_train)
neigh_predict=neigh.predict(X_test)
print("accuracy Score (training) for KNeighborsClassifier:{0:6f}".format(neigh.score(X_test,y_test)))
print("Confusion Matrix for KNeighborsClassifier:")
print(confusion_matrix(y_test,neigh_predict))
print(classification_report(neigh_predict, y_test))
```

KNN works by computing the similarity between the new data point and the rest of the data points in the dataset and classifying the new data point to the class which majority of the k nearest neighbors to the new data point belong to. Thus the training time increases linearly as the dataset increases, which increases the compuatational complexity. This why KNN does not work well with large datasets. Also, KNN does not work well when there are many features. The distance or the similarity between the data points are less meaningful when the number of features are very large. This leads to degraded performance.

## Stacking the models

```python
models = [ MLPClassifier(max_iter = 100), SVC(max_iter = 500), DecisionTreeClassifier(), RandomForestClassifier(), KNeighborsClassifier(n_neighbors=3)]

S_Train, S_Test = stacking(models,
                           X_train, y_train, X_test,
                           regression=False,

                           mode='oof_pred_bag',

                           needs_proba=False,

                           save_dir=None,

                           metric=accuracy_score,

                           n_folds=4,

                           stratified=True,

                           shuffle=True,

                           random_state=0,

                           verbose=2)
```

```python
np.savetxt("s_train.csv", S_Train, delimiter=",")
```

```python
np.savetxt("s_test.csv", S_Test, delimiter=",")
```

## Meta Model - MLP

```python
model = MLPClassifier(max_iter = 100)

model = model.fit(S_Train, y_train)
y_pred = model.predict(S_Test)
print('Final prediction score for ensemble methods: [%.8f]' % accuracy_score(y_test, y_pred))
print(classification_report(y_pred, y_test))
```

```python
#Hyperparameter tuning done for MultiLayer Perceptron classifier

parameters = {
              'activation':['logistic', 'relu'],
              'learning_rate':['constant', 'adaptive'],
              'max_iter' :[100, 150]
             }


mlp_random = RandomizedSearchCV(mlp,parameters,n_iter=15)
mlp_random.fit(y_train, y_train)
grid_parm=mlp_random.best_params_
print(grid_parm)

#Using the parameters obtained from HyperParameterTuning in the MLPClassifier
mlp = MLPClassifier(**grid_parm)
mlp.fit(X_train,y_train)
mlp_predict = mlp.predict(y_test)

#Obtain accuracy ,confusion matrix,classification report and AUC values for the result above.
print("accuracy Score (training) after hypertuning for MultiLayer Perceptron:{0:6f}".format(mlp.score(X_test,y_test)))
print("Confusion Matrix after hypertuning for MultiLayer Perceptron")
print(confusion_matrix(y_test,mlp_predict))
print("=== Classification Report ===")
print(classification_report(y_test,mlp_predict))

#get cross-validation report
mlp_cv_score = cross_val_score(mlp, X_train, y_train, cv=10, scoring="roc_auc")
print("=== All AUC Scores ===")
print(mlp_cv_score)
print('\n')
print("=== Mean AUC Score ===")
print("Mean AUC Score - MultiLayer Perceptron: ",mlp_cv_score.mean())
```
